```python
import pandas as pd 
dataset = pd.read_csv('Emotion_classify_Data.csv')
dataset.shape
```

```python
dataset
```

```python
dataset['Emotion'].value_counts()
```

```python
dataset['Emotion_num'] = dataset['Emotion'].map({
  'anger' : 0,
  'joy' : 1,
  'fear' : 2
})
```

```python
X = dataset['Comment']
y = dataset['Emotion_num']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21, stratify=y)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier

rfp = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()), 
     ('Random Forest', RandomForestClassifier())         
])

rfp.fit(X_train, y_train)
```

```python
y_pred = rfp.predict(X_test)

from sklearn.metrics import classification_report
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
from sklearn.naive_bayes import MultinomialNB

multip = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),    
     ('Multi NB', MultinomialNB())         
])

multip.fit(X_train, y_train)
```

```python
y_pred2 = multip.predict(X_test)

cr = classification_report(y_test, y_pred2)
print(cr)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knnp = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),    
     ('KNN', KNeighborsClassifier())         
])

knnp.fit(X_train, y_train)
```

```python
y_pred3 = knnp.predict(X_test)

cr = classification_report(y_test, y_pred3)
print(cr)
```

```python
import spacy

nlp = spacy.load('en_core_web_sm')

def preperocess(text):
  doc = nlp(text)
  filtered_tokens = []

  for token in doc:
    if token.is_stop and token.is_punct:
      continue
    filtered_tokens.append(token.lemma_)

  return ' '.join(filtered_tokens)
```

```python
dataset['preprocessed_data'] = dataset['Comment'].apply(preperocess)
```

```python
X = dataset['preprocessed_data']
y = dataset['Emotion_num']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=21, stratify=y)
```

```python
rfp = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()), 
     ('Random Forest', RandomForestClassifier())         
])

rfp.fit(X_train, y_train)
```

```python
y_pred = rfp.predict(X_test)

from sklearn.metrics import classification_report
cr = classification_report(y_test, y_pred)
print(cr)
```

```python
from sklearn.naive_bayes import MultinomialNB

multip = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),    
     ('Multi NB', MultinomialNB())         
])

multip.fit(X_train, y_train)
```

```python
y_pred2 = multip.predict(X_test)

cr = classification_report(y_test, y_pred2)
print(cr)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knnp = Pipeline([
     ('vectorizer_tfidf',TfidfVectorizer()),    
     ('KNN', KNeighborsClassifier())         
])

knnp.fit(X_train, y_train)
```

```python
y_pred3 = knnp.predict(X_test)

cr = classification_report(y_test, y_pred3)
print(cr)
```
